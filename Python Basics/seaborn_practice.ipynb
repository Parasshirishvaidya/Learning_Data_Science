```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
# Depedent feature  : - Tip
# Independent Feature : - total_bill,sex,smoker,day,time,size
```

# Correlation with Heatmap
A correlation heatmap uses coloured sells, typically in monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
df[['total_bill','tip','size']].corr()
```

```python
sns.heatmap(df[['total_bill','tip','size']].corr())
```

# Bivariate Analysis

```python
sns.jointplot(x='tip',y='total_bill', data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill', data=df,kind='reg')
```

# Pair Plot

A pair plot is also known as a scatterplot, in which one varaible in the smae data row is matched with another variable's value, like this: Pair plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue='smoker')
```

# Displot

```python
sns.distplot(df['total_bill'])
```

# Categorical Plots

Seaborn also helps us in doing the analysis on categorical data point.

```python
## Count Plot
sns.countplot(x='day',data=df)
```

```python
sns.barplot(x='total_bill',y='smoker',data=df)
```

# Boxplot

```python
sns.boxplot(x='day', y='total_bill',hue='smoker', data=df,palette='rainbow')
```

# Violin Plot

```python
sns.violinplot(x='total_bill', y="day", data=df, palette='rainbow')
```
