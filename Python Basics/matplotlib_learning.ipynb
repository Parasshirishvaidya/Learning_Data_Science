## Matplotlib

matplotlib is a plotting library for the python programming language and its numerical mathermatics extention NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Ot or GTK+

Some of the major Pros of Matplotlib are: -
1) Generally easy to get started for simple plots
2) Support for custom labels and texts
3) Greate control fo every element in a figure
4) High-quality output in many formants
5) Very Customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
x=np.arange(40,50)
y=np.arange(50,60)
```

```python
# Using scatter Plot 
## show data points on a graph

plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
# Simple example 
x=np.arange(0,10)
y=np.arange(11,21)
y=x*x
plt.plot(x,y,'r*--')
```

```python
## Creating Subplots ie many plots in a single canvas 

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

```python
# Compute the x and y coordinates for points on a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("Sine wave form")
plt.plot(x,y)
```

```python
#Compute the x and y coordinates for points on sine and Cosine Curves
x = np.arange(0,5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#create sub plt grid
plt.subplot(2,1,1)

#Make first plot
plt.plot(x,y_sin)
plt.title('Sine')

#Second subplot
plt.subplot(2,1,2)
plt.plot(x,y_cos, c='y')
plt.title('Cosine')
```

# Bar plot

# Histogram

```python
a=np.random.randint(0,100,size=20)
plt.hist(a)
plt.title("Histogram")
plt.show()
a
```

# Box Plot 

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
```

# Pie Chart

```python
# Data to plot
labels = ['Python','C++','Ruby','Java']
sizes = [215,146,123,246]
colors = ["gold","yellowgreen","LightCoral","lightskyblue"]
explode = (0.1,0.1,0.1,0.1) # explode ist slice

# plot
plt.pie(sizes,explode=explode, labels=labels,colors=colors,
       autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
