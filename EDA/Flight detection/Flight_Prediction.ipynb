## Flight Price Prediction (EDA and Feature Engineering)

```python
# importing basic librariies 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel('Data_train.xlsx')
train_df.head()
```

```python
train_df.info()
```

```python
test_df=pd.read_excel('test_set.xlsx')
test_df.head()
```

```python
df = pd.concat([train_df, test_df], ignore_index=True)
df.head()
```

```python
df.isnull().sum()
```

```python
## EDA and Feature engineering of date of journey
# we can create a day month and year features from Data of journey feature 
# these are called derived feature 
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
## Arrival Time Feature engineering 
df['Arrival_Time']=df['Arrival_Time'].str.split(" ").str[0]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Arrival_hour']=df['Arrival_Time'].apply(lambda x : x.split(':')[0])
df['Arrival_min']=df['Arrival_Time'].apply(lambda x : x.split(':')[1])
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Dep_hour']=df['Dep_Time'].apply(lambda x : x.split(':')[0])
df['Dep_min']=df['Dep_Time'].apply(lambda x : x.split(':')[1])
```

```python
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
df.head()
```

```python
df.drop('Route',axis=1,inplace=True)
```
